```python
import os
import requests

# URL base e lista de arquivos
base_url = "http://vitibrasil.cnpuv.embrapa.br/download/"
files = [
    "Producao.csv", "ProcessaViniferas.csv", "ProcessaAmericanas.csv",
    "ProcessaMesa.csv", "ProcessaSemclass.csv", "Comercio.csv",
    "ImpVinhos.csv", "ImpEspumantes.csv", "ImpFrescas.csv",
    "ImpPassas.csv", "ImpSuco.csv", "ExpVinho.csv",
    "ExpEspumantes.csv", "ExpUva.csv", "ExpSuco.csv"
]

# Diretório onde os arquivos serão salvos
download_dir = "csv"
os.makedirs(download_dir, exist_ok=True)

# Função para verificar se a página está online e baixar o arquivo
def download_file(file_name):
    url = base_url + file_name
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            file_path = os.path.join(download_dir, file_name)
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Arquivo {file_path} baixado com sucesso.")
        else:
            print(f"Arquivo indiponivel {url}: Status code {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Erro ao tentar acessar {url}: {e}")

# Verifica todas as URLs e faz o download dos arquivos
for file in files:
    download_file(file)
```

```python
import os
import pandas as pd
import requests

# arquivos e tipos
csv_types = [
    # {"Comercio.csv": "tipo_1"},
    # {"Producao.csv": "tipo_1"},
    # {"ProcessaAmericanas.csv": "tipo_2"},
    # {"ProcessaMesa.csv": "tipo_2"},
    # {"ProcessaSemclass.csv": "tipo_2"},
    # {"ProcessaViniferas.csv": "tipo_2"},
    {"ImpVinhos.csv": "tipo_3"},
    # {"ImpEspumantes.csv": "tipo_3"},
    # {"ImpFrescas.csv": "tipo_3"},
    # {"ImpPassas.csv": "tipo_3"},
    # {"ImpSuco.csv": "tipo_3"},
    # {"ExpVinho.csv": "tipo_3"},
    # {"ExpEspumantes.csv": "tipo_3"},
    # {"ExpUva.csv": "tipo_3"},
    # {"ExpSuco.csv": "tipo_3"}
]

# Função ajustada para usar o campo 'Produto' em vez de 'control'
def fill_item_column_from_produto(df):
    current_item = None
    item_column = []
    
    for produto_value in df['subitem']:
        # Verifica se o valor é uma string antes de tentar verificar se está em maiúsculas
        if isinstance(produto_value, str) and produto_value.isupper():
            current_item = produto_value
        # Adiciona o item atual à coluna
        item_column.append(current_item)
    
    return item_column

for entry in csv_types:
    for csv_name, csv_type in entry.items():

        if csv_type == 'tipo_1':
            # Carregar o arquivo usando o delimitador correto ";"
            df = pd.read_csv('csv/' + csv_name, sep=";")
            
            # Remover as colunas 'id' e 'control'
            df.drop(columns=['id', 'control'], inplace=True)
            
            # Renomeando coluna
            df.rename(columns={'Produto': 'subitem'}, inplace=True)
            
            # Aplicar a função para preencher a nova coluna ITEM usando o campo 'Produto'
            df['item'] = fill_item_column_from_produto(df)            

            # Reordenar as colunas, colocando 'ITEM' na primeira posição
            cols = ['item'] + [col for col in df.columns if col != 'item']
            df = df[cols]

            # Processo de Unpivot
            # Identificar colunas que representam os anos (colunas que são números)
            year_columns = [col for col in df.columns if col.isdigit()]

            # Fazer o melt para transformar as colunas de anos em linhas
            df_melted = pd.melt(df, id_vars=['item', 'subitem'], value_vars=year_columns,
                                var_name='ano', value_name='valor')
        
        if csv_type == 'tipo_2':
            # Carregar o arquivo usando o delimitador correto ";"
            df = pd.read_csv('csv/' + csv_name, sep="\t")
            
            # Remover as colunas 'id' e 'control'
            df.drop(columns=['id', 'control'], inplace=True)
            
            # Renomeando coluna
            df.rename(columns={'cultivar': 'subitem'}, inplace=True)
            
            # Aplicar a função para preencher a nova coluna ITEM usando o campo 'Produto'
            df['item'] = fill_item_column_from_produto(df)
            
            # Reordenar as colunas, colocando 'ITEM' na primeira posição
            cols = ['item'] + [col for col in df.columns if col != 'item']
            df = df[cols]

            # Processo de Unpivot
            # Identificar colunas que representam os anos (colunas que são números)
            year_columns = [col for col in df.columns if col.isdigit()]

            # Fazer o melt para transformar as colunas de anos em linhas
            df_melted = pd.melt(df, id_vars=['item', 'subitem'], value_vars=year_columns,
                                var_name='ano', value_name='valor')
            
        if csv_type == 'tipo_3':
            # Carregar o arquivo usando o delimitador correto ";"
            df = pd.read_csv('csv/' + csv_name, sep=";")
            
            # Renomeando coluna
            #df.rename(columns={'País': 'paises'}, inplace=True)
            
            # Identificar as colunas de kg e valor
            kg_cols = [col for col in df.columns if not '.1' in col and col not in ['Id', 'País']]
            valor_cols = [col for col in df.columns if '.1' in col]

            # Renomear as colunas de valor para remover o ".1"
            valor_cols_renamed = [col.replace('.1', '') for col in valor_cols]

            # Reorganizar os dados de kg e valor em DataFrames separados
            df_kg = df[['Id', 'País'] + kg_cols]
            df_valor = df[['Id', 'País'] + valor_cols]

            # Renomear as colunas de valor para que correspondam às de kg
            df_valor.columns = ['Id', 'País'] + valor_cols_renamed

            # Usar melt para transformar os dados de kg e valor separadamente
            df_kg_melted = df_kg.melt(id_vars=['Id', 'País'], var_name='Ano', value_name='kg')
            df_valor_melted = df_valor.melt(id_vars=['Id', 'País'], var_name='Ano', value_name='valor')

            # Combinar os dois DataFrames com base em 'Id', 'País', e 'Ano'
            df_final = pd.merge(df_kg_melted, df_valor_melted, on=['Id', 'País', 'Ano'])

            # Remover duplicatas e manter as colunas necessárias
            df_final = df_final[['País', 'Ano', 'kg', 'valor']]

            # Ordenar o DataFrame pelo Id e Ano
            #df_final = df_final.sort_values(by=['Id', 'Ano']).reset_index(drop=True)

            # Exibir o DataFrame final
            print(df_final.head())
```
